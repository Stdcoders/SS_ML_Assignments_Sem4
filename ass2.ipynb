```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

# Bar Plot for age distribution 

```python

ins['age_bins'] = pd.cut(ins['age'], bins=range(10, 81, 10))
age_barplot = ins['age_bins'].value_counts().sort_index()
age_barplot.plot(kind='bar', color='skyblue')
plt.title("Age distribution")
plt.xlabel("Age Groups")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

# Scatter plot for smoker and insurance charges

```python
sns.scatterplot(data=ins, x='age', y='charges', hue='smoker', alpha=0.7)
plt.title("Age vs Insurance Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.legend(title="Smoker")
plt.show()
```

# pie chart for gender distribution

```python
ins['sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['pink', 'lightgreen'])
plt.title("Gender Proportion")
plt.show()
```

# box plot for BMI distribution 

```python
sns.boxplot(data=ins, y='bmi',palette='pastel')
plt.title("BMI Distribution")
plt.ylabel("BMI")
plt.show()
```

# regional proportion using pie chart

```python
ins['region'].value_counts().plot(kind='pie', autopct='%1.1f%%', colormap='coolwarm')
plt.title("Regional Proportion")
plt.ylabel("")
plt.show()
```

# pair plot for all features using smoker as hue

```python
sns.pairplot(ins, hue='smoker', diag_kind='kde', palette='Set2')
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

# histogram for distribution of charges

```python
ins['charges'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Charges")
plt.xlabel("Charges")
plt.ylabel("Frequency")
plt.show()
```
